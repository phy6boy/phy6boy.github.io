# The Central Limit Theorem
The central limit theorem is an important theorem in statistics.

```python
import numpy as np
import scipy as sp
import scipy.optimize as op
from scipy import stats
import matplotlib.pyplot as plt
```

```python
def gauss(x, A, B, x0):
    y = A*np.exp(-1*B*(x-x0)**2)
    return y
```

```python
ntrial = 500
nsim = 10000
x = np.sum(np.random.choice(2, (ntrial, nsim)), axis=0)
hist, edge = np.histogram(x, bins=ntrial, density=True)
edge = (edge[:-1] + edge[1:])/2
mask = hist!=0
hist = hist[mask]
edge=edge[mask]
p, cov = op.curve_fit(gauss, edge, hist, p0=[0.0,1.0,ntrial/2])
x = np.linspace(min(edge), max(edge), 1000)
y = gauss(x, p[0], p[1], p[2])
plt.bar(edge, hist)
plt.plot(edge, hist, "or")
plt.plot(x, y, "k--", lw=2.5)
plt.show()
```
